Current date loan sum

```python
curr_date = pd.Timestamp.today()

curr_month = curr_date.month

curr_day = curr_date.day

no = df[(df['issue_date'].dt.month == curr_month) & (df['issue_date'].dt.day <= curr_day)]

sum = no['loan_amount'].sum()

amt_millions = sum / 1000000

amt_in_millions = round(amt_millions, 2)

print(f'Total amount invested: ${amt_in_millions} M')
```

```python
loan_status_data = df.groupby('loan_status').agg({
    'id' : 'count',
    'dti' : 'mean',
    'int_rate' : 'mean',
    'loan_amount' : 'sum',
    'total_payment' : 'sum'
})

loan_status_data.rename(columns={'id' : 'No of Applications'}, inplace=True)

display(loan_status_data)
```

Good Loan Analysis

```python
# No of Applications

good_loan_app = loan_status_data.loc['Fully Paid' , 'No of Applications'] + loan_status_data.loc['Current' , 'No of Applications']

print(good_loan_app)


# Funded Amount

funded_amount = loan_status_data.loc['Fully Paid' , 'loan_amount'] + loan_status_data.loc['Current' , 'loan_amount']

print(funded_amount)

# Received Amount

receiver_amt = loan_status_data.loc['Fully Paid' , 'total_payment'] + loan_status_data.loc['Current' , 'total_payment']

print(receiver_amt)

# Profit

print(receiver_amt - funded_amount)
```

Bad Loan Analysis

```python
# No of Applications

bad_loan_app = loan_status_data.loc['Charged Off', 'No of Applications']

print(bad_loan_app)

# Funded Amount

funded_amount = loan_status_data.loc['Charged Off' , 'loan_amount']

print(funded_amount)

# Received Amount

receiver_amt = loan_status_data.loc['Charged Off' , 'total_payment']

print(receiver_amt)

# Profit

print(receiver_amt - funded_amount)
```
