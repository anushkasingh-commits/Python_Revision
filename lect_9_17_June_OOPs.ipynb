```python
class student:

  # Constructor

  def __init__(self, name, age):
    self.name = name
    self.age = age

  def display(self):
    print('Name: ', self.name)
    print('Age: ', self.age)

a = student('John', 21)

# print(a.name)
# print(a.age)

a.display()
```

```python
from os import name
class person:
  def __init__(self, h, w, age):

    self.height: h
    self.weight = w
    self.age = age

class Rahul(person):

  def __init__(self, g, name, b, h, w, age):

    super().__init__(h, w, age)

    self.gender = g
    self.name = name
    self.bank_bal = b

r = Rahul('Male', 'Rahul', 5000, 5, 55, 10)
print(r.age)
```

```python
class complex:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def __add__(self, other):
    return complex(self.x + other.x, self.y + other.y)

  def __str__(self):
        return f"{self.x} + i{self.y}"

a = complex(2, 3)
b = complex(4, 5)

c = a + b
print(c.x, c.y)
```
