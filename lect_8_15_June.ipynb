```python
import pandas as pd
```

```python
# Inner Join

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.merge(df1, df2, on = 'id', how = 'inner')

df3.rename(columns = {'name_x' : 'table_name1'}, inplace = True)

print(df3)
```

```python
# Left Join

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.merge(df1, df2, on = 'id', how = 'left')

# df3.rename(columns = {'name_x' : 'table_name1'}, inplace = True)

print(df3)
```

```python
# Right Join

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.merge(df1, df2, on = 'id', how = 'right')

# df3.rename(columns = {'name_x' : 'table_name1'}, inplace = True)

print(df3)
```

```python
# Outer Join

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.merge(df1, df2, on = 'id', how = 'outer')

# df3.rename(columns = {'name_x' : 'table_name1'}, inplace = True)

print(df3)
```

```python
# Cross Join

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.merge(df1, df2, how = 'cross')

# df3.rename(columns = {'name_x' : 'table_name1'}, inplace = True)

print(df3)
```

```python
# Concat function

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.concat([df1, df2])

print(df3)
```

```python
# drop_duplicate

df1 = pd.DataFrame({
    'id' : [1, 2, 3, 4],
    'name' : ['Anushka', 'Rahul', 'Raj', 'Riya'],
    'age' : [21, 23, 56, 89]
})
df2 = pd.DataFrame({
    'id' : [3, 4, 5, 6],
    'name' : ['Ravi', 'Rahul', 'Surya', 'Raj'],
    'age' : [23, 40, 50, 89]
})
df3 = pd.concat([df1, df2])

df3 = df3.drop_duplicates(subset = ['id', 'name'], keep = 'first')   # , inplace = True (another parameter)

print(df3)
```
