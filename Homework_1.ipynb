```python
import numpy as np
```

```python
scores = []
days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
for i in range(7):
  x = int(input(f'Scores on {days[i]}: '))
  scores.append(x)
print(scores)
arr = np.array(scores)
max_score = np.argmax(scores)
min_score = np.argmin(scores)

print(f"Max score is {arr[max_score]} on {days[max_score]}")
print(f"Min score is {arr[min_score]} on {days[min_score]}")
print(f"Total score is {np.sum(arr)}")
print(f"Standard deviation is {np.std(arr)}")
print(f"Mean is {np.mean(arr)}")
```

```python
price = []
items = ['Sugar', 'Salt', 'Rice', 'Flour']
for i in range(4):
  x = int(input(f"Price of {items[i]}: "))
  price.append(x)
print(price)

arr = np.array(price)
max_price = np.argmax(price)
min_price = np.argmin(price)

print(f"Max price is {arr[max_price]} of {items[max_price]}")
print(f"Min price is {arr[min_price]} of {items[min_price]}")
print(f"Total price is {np.sum(arr)}")
print(f"Mean is {np.mean(arr)}")
print(f"Standard deviation is {np.std(arr)}")
```
